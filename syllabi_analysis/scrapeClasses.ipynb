```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
urls = [
    "http://coursecatalog.web.cmu.edu/schools-colleges/tepper/undergraduatebusinessadministrationprogram/courses/",
    "http://coursecatalog.web.cmu.edu/schools-colleges/tepper/undergraduateeconomicsprogram/courses/"
]

degree = ['BUS', 'ECO']
```

```python
classes = []
semesters = []
for i in range(2):
    res = requests.get(urls[i])
    htmlData = res.content
    parsedData = BeautifulSoup(htmlData, "html.parser")
    for entry in parsedData.find_all("dt"):
        className = entry.getText()
        courseID = className[:6].split('-')
        courseID = ''.join(courseID)
        courseName = className[7:]
        classes.append([courseID, courseName, degree[i]])
    for entry in parsedData.find_all("dd"):
        br_tag = entry.find('br')
        if br_tag:
            semester_text = br_tag.previous_sibling.strip()
        else:
            semester_text = entry.get_text(strip=True)  # If no <br>, take entire text
        semesters.append(semester_text.split(':')[0])
classes[:10]
```

```python
semesters[-10:]
```

```python
classdf = pd.DataFrame(classes, columns = ['CourseID', 'CourseName','DegreeProgram'])

classdf.head()
```

```python
# add semester offered

classdf['SemesterOffered'] = semesters
classdf.head()
```

```python
classdf.to_csv("scraped_courses.csv")
```
